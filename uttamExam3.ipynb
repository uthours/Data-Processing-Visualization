#c) Create a histogram of the ‘Goals’.

```python
ax = df.Goals.hist()
plt.show()
```

```python
ax = df.Goals.hist(bins = 25)
ax.set(title = "Goals") #creating title
ax.set_xlabel("Number of Goals", labelpad=20, size=12)
ax.set_ylabel("Number of Players", labelpad=20, size=12)
plt.show()
```

#d) Create a histogram of the ‘Goals’ and ‘Assists’ on the same axis.

```python
df[['Goals','Assists']].plot.hist(alpha = 0.5, bins = 25)
```

It is intresting how the number of assists are so high. Them players tried really hard to score it

#e) Create a barchart for most goals scored by nationality.

```python
ax = df.groupby('nationality').Goals.agg('mean').plot(kind = 'bar')
plt.show()
```

Egypt has scored the most goals and Costa Rica has least number of goals.

```python
ax = df.groupby('nationality').Goals.agg('mean').plot(kind = 'bar')
ax.set(title = "Most Goals  by Nationality")
ax.set(ylabel = 'Number of goals')
plt.show()
```

#f) Create a boxplot of ‘Yel’ (yellow cards) by nationality.

```python
df.boxplot(column = 'Yel', by = 'nationality', rot = 90)
plt.show()
```

Columbia has the most yellow cards and Japan has the least number of yellow cards.

#g) Create ‘Goals’ and ‘Assists’ as scatter. Signify your favorite club or nationality in some way on the scatterplot. You may need to add a new column to the dataset to do so

I am going to plot the goals and assist in the dataset then stacked graph by nationality

```python
sns.relplot(data=df, x = 'Goals', y= 'Assists' )
plt.title('A scatterplot of goals and assists')
```

```python
dfbar = df.groupby(['nationality'])['Goals'].sum()
dfbar.plot(kind='bar', stacked="True", title= "Stacked Graph of Goals by Nationality")
plt.show()
```

#h) Create stacked barplot of goals and assists of French National Team. (they won this tourney!) Be sure to comment on why it is appropriate to stack goals and assists.

```python
df3 = df[df['nationality'].str.contains('France')]
```

```python
ax = df3[['Goals', 'Assists']].plot(kind = 'bar', stacked = True,figsize=(15, 10) , title = 'French National Team')
ax.set_xlabel('Goals')
ax.set_ylabel('Assists')
```

The barplot shows the goals and assists of French national team. The game is based on highest number of goals by a team, but assists are also as important as goals. The goals and assists are the major counts in the game so it is appropriate to plot them in a graph for better visualization. 

#i) Create a correlation heatmap of the entire dataset.

```python
ax = sns.heatmap(df.corr(),vmin = -1)
ax.set(title = 'Heatmap of Football')

plt.show()
```
