Now that I am authenticated, I can start to play around in the dataset.  I am going to look at the Austin bikeshare trips data from Texas and calculate the entries in the dataset, compute the average time for a trip based on starting point and compute how many trips start at each starting point.

```python
%%bigquery --project fabled-progress-338823
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5
```

I am going to plot the station name and average time in the following code

```python
%%bigquery --project fabled-progress-338823
SELECT AVG(duration_minutes) as avg_time, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
Order by avg_time
```

lets put the data related to no_trips and start station name.

```python
%%bigquery --project fabled-progress-338823
SELECT COUNT(start_station_name) as No_Trip, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
Order by No_Trip
```
